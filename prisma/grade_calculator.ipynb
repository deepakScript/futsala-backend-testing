```python
class Grade:
    def __init__(self,sub, *marks):
        self.sub = sub
        self.marks = marks
    
    def getMarks(self):
        print(f"Enter the number of subjects:")
        self.sub = int(input())
        for i in range(self.sub):
            mark = int(input(f"Enter marks for subject {i+1}: "))
            self.marks += (mark,)

    def total_marks(self):
        print(f"Total marks obtained: {sum(self.marks)}")
    
    def percentage(self):
        total = int(sum(self.marks))
        print(f"Enter total percentage {total/(self.sub * 100) * 100}%") 


student = Grade(0)
student.getMarks()
student.total_marks()
student.percentage()

```

```python
class Employee:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def input_details(self):
        self.name = input("Enter employee name: ")
        self.age = int(input("Enter employee age: "))

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


    

class PartEmpoyee(Employee):
    def __init__(self, name, age, hourly_rate, hours_worked):
        super().__init__(name, age)
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def input_details(self):
        super().input_details()
        self.hourly_rate = int(input("Enter part time employee hourly rate: "))
        self.hours_worked = int(input("Enter part time employee hours worked: "))

    def display(self):
        super().display()
    
    def calculate_salary(self):
        total_salary = self.hourly_rate * self.hours_worked
        print(f"Part time total salary: {total_salary}")


class FullEmpoyee(Employee):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def input_salary(self):
        self.salary = int(input("Enter full time employee salary: "))

    def display(self):
        super().display()
        print(f"Full time total salary: {self.salary}")


# part_time_emp = PartEmpoyee(0, 0, 0, 0)
# full_time_emp = FullEmpoyee()
# part_time_emp.input_details()
# full_time_emp.input_details()
# part_time_emp.display()
# part_time_emp.calculate_salary()
# full_time_emp.display()

full_time_emp = FullEmpoyee(0, 0, 0)
full_time_emp.input_details()
full_time_emp.input_salary()
full_time_emp.display()
```
